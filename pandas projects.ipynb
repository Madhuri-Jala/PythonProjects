```python
import numpy as np
import pandas as pd
#creating the random forces and areas
forces_x = np.random.randint(1000, 5000, 20)
forces_y= np.random.randint(1000,5000,20)
forces_z=np.random.randint(1000,5000,20)
areas = np.random.randint(10, 50, 20)
#creating dataframe
df=pd.DataFrame({"Fx":forces_x,"Fy":forces_y,"Fz":forces_z,"areas": areas})
#calculting the stress
df["stress_x"]=df["Fx"]/df["areas"]
df["stress_y"]=df["Fy"]/df["areas"]
df["stress_z"]=df["Fz"]/df["areas"]
print("stress,forces,areas 3d table:\n",df)
#saving to excel file
a=df.to_excel("C://Users//MADHURI//Desktop//stress_data.xlsx", index=False)
print("saved to excel")
#info
b=df.info()
print("------info of the stress table-----",b)
#Failure Detection System stress_x
df.loc[df["stress_x"] > 150,"Failure Detection stress_x"] = "Failure"
print("\n updateed table :\n",df)
#Failure Detection System stress_y
df.loc[df["stress_y"] > 150,"Failure Detection stress_y"] = "Failure"
print("\n updateed table :\n",df)
#Failure Detection System stress_z
df.loc[df["stress_z"] > 150,"Failure Detection stress_z"] = "Failure"
print("\n updateed table :\n",df)
#Safe Detection System
df.loc[df["stress_x"] < 150,"Failure Detection stress_x"] = "Safe"
df.loc[df["stress_y"] < 150,"Failure Detection stress_y"] = "Safe"
df.loc[df["stress_z"] < 150,"Failure Detection stress_z"] = "Safe"
print("\n updateed table :\n",df)
#max stress
df["max_total_stress"]=(df["stress_x"]**2+df["stress_y"]**2+df["stress_z"]**2)**0.5
max_stress=df["max_total_stress"].max()
print("max_total_stress:\n",max_stress)
```

- Stress is computed using combined force components and area.
- Stress ratio indicates how close the structure is to failure.
- Safety factor provides a measure of structural safety margin.
- Failure occurs when stress exceeds yield strength.
- Dataset is now ready for machine learning applications.

```python
import numpy as np
import pandas as pd
#number of sample= 50
n=50
#dataframes of forces,areas andcomponents
loads_df = pd.DataFrame({
    "Component": np.random.choice(["Wing", "Fuselage", "LandingGear"], n),
    "Fx": np.random.randint(1000, 5000, n),
    "Fy": np.random.randint(1000, 5000, n),
    "Fz": np.random.randint(1000, 5000, n),
    "Area": np.random.randint(10, 50, n)
})
print("Loads:\n", loads_df.head())
#loads data
loads_df["Stress"] = (
    np.sqrt(loads_df["Fx"]**2 + loads_df["Fy"]**2 + loads_df["Fz"]**2)
) / loads_df["Area"]
print("\nLoads with Stress:\n", loads_df.head())
# material dataframe of matrials name, Yield_Strength, density
materials_df = pd.DataFrame({
    "Component": ["Wing", "Fuselage", "LandingGear"],
    "Material": ["Aluminum", "Titanium", "Steel"],
    "Yield_Strength": [300, 900, 600],
    "Density": [2700, 4500, 7850]
})
print("\nMaterials:\n", materials_df)
#check for duplicates
print("\nDuplicates in loads:", loads_df.duplicated().sum())
print("Duplicates in materials:", materials_df.duplicated().sum())
#merge_df both matrials_df and loads_df
merged_df = pd.merge(loads_df, materials_df, on="Component")
print("\nMerged Data:\n", merged_df.head())
#stress ratio
merged_df["Stress_ratio"]=merged_df["Stress"] / merged_df["Yield_Strength"]
print("merged_df stress_ratio:\n",merged_df.head())
#Safety factor
merged_df["Safety_factor"]=merged_df["Yield_Strength"] / merged_df["Stress"]
print("merged_df of safety factor:\n",merged_df.head())
#Normalize stress data
merged_df["Stress_norm"] = (
    (merged_df["Stress"] - merged_df["Stress"].min()) /
    (merged_df["Stress"].max() - merged_df["Stress"].min()))

print("normalize stress data:\n",merged_df["Stress_norm"].head())
#failure
merged_df["Failure"] = merged_df["Stress_ratio"] > 1
print("failure:\n",merged_df["Failure"].head())
```

```python
# - Free-stream velocity shows strong influence on sound pressure level.
# - Angle of attack also significantly impacts aerodynamic noise.
# - Some parameters show weaker correlation, indicating lesser influence.
import numpy as np
import pandas as pd
d=pd.read_csv("/content/drive/MyDrive/NASA_airfoil_self_noise.csv")
airfoil_df=pd.DataFrame(d)
#knowing the columns
print(airfoil_df.columns)
#Feature correlations
corr_df=airfoil_df.corr()
print(corr_df)
#target name
target="Sound"
#correlation with target
corr_target=corr_df[target].sort_values(ascending=False)
print("corr_target:\n",corr_target)
#important features
important_features = corr_target[1:]
print("important targets:\n",important_features)

```

- Sudden changes in velocity and acceleration were identified as anomalies.
- Large positive changes indicate spikes, while large negative changes indicate drops.
- These anomalies may represent instability, turbulence, or system irregularities.
- Time-series analysis helps in monitoring UAV performance and detecting abnormal behavior.


```python
import numpy as np
import pandas as pd
n=10
Altitudes=np.random.uniform(1000,5000,n)
Velocity=np.random.uniform(350,700,n)#km/h
Acceleration=np.random.uniform(1,3.5,n)#m/s*2
Time=np.arange(n)#hr
#UAV Flight Data
Flight_Data=pd.DataFrame({"Altitudes":Altitudes,"Velocity":Velocity,"Acceleration":Acceleration,"Time":Time})
print(Flight_Data)
#Altitudes,Velocity changes
Flight_Data["Acceleration_change"]=Flight_Data["Acceleration"].diff()
Flight_Data["Velocity_change"]=Flight_Data["Velocity"].diff()
print("Velocity_change:\n",Flight_Data["Velocity_change"])
print("Acceleration_change:\n",Flight_Data["Acceleration_change"])
#Define thresholds
velocity_threshold = 100
acceleration_threshold = 1
#flight_data_altitudes_anomalies
Flight_Data["Acceleration_Anomalies"]= (Flight_Data["Acceleration_change"] > acceleration_threshold)
print("Acceleration_Anomalies:\n",Flight_Data["Acceleration_Anomalies"])
Flight_Data["Velocity_Anomalies"]=(Flight_Data["Velocity_change"]>velocity_threshold)
print("Velocity_Anomalies:\n",Flight_Data["Velocity_Anomalies"])

```

- Failure occurs when applied stress exceeds yield strength.
- Higher loads and smaller areas increase stress, leading to failure.
- Safety factor < 1 indicates unsafe conditions.
- Dataset contains both safe and failure cases, making it suitable for machine learning.



```python
import numpy as np
import pandas as pd
n=5
loads=np.random.randint(5000,8000,n)
areas=np.random.uniform(0.5,20,n)
Yield_Strength=np.array([450,1000,550,1300,500])
materials=np.array(["Aluminum Alloy (2024 / 7075)","Titanium Alloy","Steel",
                     "Carbon Fiber Reinforced Polymer","Glass Fiber Reinforced Polymer"])
Structural_df=pd.DataFrame({"loads":loads,"areas":areas,"Yield_Strength":Yield_Strength,"materials":materials})
Structural_df["stress"]=Structural_df["loads"]/Structural_df["areas"]
#computational stress
print("Stress:\n",Structural_df["stress"])
#failure condition
Structural_df["failure"]=(Structural_df["stress"]>Structural_df["Yield_Strength"]).astype(int)
print('failure of metireal:\n',Structural_df["failure"])
#safety factor
Structural_df["Safety_Factor"] = Structural_df["Yield_Strength"] / Structural_df["stress"]
print(Structural_df["Safety_Factor"])
print(Structural_df)
```
