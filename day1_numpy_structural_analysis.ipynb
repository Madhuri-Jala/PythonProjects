**project-2** Failure Detection
Find all values where stress > 120
Print their indices
Count how many failure points exist

```python
import numpy as np
forces=np.random.randint(1000,5000,20)
areas=np.random.randint(5,25,20)
stress= forces/areas
failure_values=stress[stress>110]
indices_points=np.where(stress>110)
print("failure_values:\n",failure_values)
print("indices points:\n",indices_points)
```

**PROBLEM 4** — 2D Structural Data
Reshape your stress into a 4×5 matrix
Print matrix
Find mean along:
axis=0
axis=1

```python
import numpy as np
forces=np.random.randint(1000,5000,20)
areas=np.random.randint(5,25,20)
stress=forces/areas
A=stress.reshape(4,5)
print("reshape into 2D:\n",A)
B=np.mean(A,axis=0)
print("mean of stress along axis=0:\n",B)
C=np.mean(A,axis=1)
print("mean of stress along axis=1:\n",C)
```

**PROBLEM 5 **— Beam Comparison (IMPORTANT)
Create 2 beams:
Beam A stress values
Beam B stress values
Stack them into 2D array
Compare which beam has higher average stress
Find difference between beams

```python
import numpy as np
beamA=np.array([[1,2],[3,4]])
beamB=np.array([[5,6],[7,8]])
C=np.stack((beamA.flatten(),beamB.flatten()))
print("stack them into 2d array:\n",C)
D=np.mean(beamA)
print(" mean of beamA:\n",D)
F=np.mean(beamB)
print(" mean of beamB:\n",F)
if D>F:
    print("beamA isis more critical")
else:
    print("beamB is is more critical")
E=np.subtract(beamA,beamB)
print("difference btw beams:\n ",E)
```
